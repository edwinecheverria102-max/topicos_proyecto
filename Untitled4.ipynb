```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
import time
import copy
import os

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Dispositivo de entrenamiento: {device}")

data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

data_dir = '/content/dataset'

try:
    image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), data_transforms[x])
                      for x in ['train', 'val']}
    dataloaders = {x: torch.utils.data.DataLoader(image_datasets[x], batch_size=32,
                                                 shuffle=True, num_workers=2)
                  for x in ['train', 'val']}
    dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val']}
    class_names = image_datasets['train'].classes
    print(f"Clases detectadas: {class_names}")
except FileNotFoundError:
    print("No se encontró la carpeta 'dataset'")

# cargar el modelo preentrenado
model_ft = models.mobilenet_v2(weights=models.MobileNet_V2_Weights.IMAGENET1K_V1)

for param in model_ft.parameters():
    param.requires_grad = False

# Reemplazar el clasificador final para que coincida con nuestro número de clases ambientales
num_ftrs = model_ft.classifier[1].in_features
num_classes = 3 if 'class_names' not in locals() else len(class_names)

model_ft.classifier[1] = nn.Sequential(
    nn.Dropout(p=0.2, inplace=False),
    nn.Linear(num_ftrs, num_classes)
)

model_ft = model_ft.to(device)

#Función de Pérdida y el Optimizador
criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.Adam(model_ft.classifier[1].parameters(), lr=0.001)

#Entrenamiento
def train_model(model, criterion, optimizer, dataloaders, dataset_sizes, num_epochs=10):
    since = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f'Época {epoch}/{num_epochs - 1}')
        print('-' * 10)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Modelo en modo entrenamiento
            else:
                model.eval()   # Modelo en modo evaluación

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f'{phase} Pérdida: {epoch_loss:.4f} Precisión: {epoch_acc:.4f}')

            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

        print()

    time_elapsed = time.time() - since
    print(f'Entrenamiento completado en {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Mejor precisión en validación: {best_acc:4f}')

    # Cargar los mejores pesos del modelo
    model.load_state_dict(best_model_wts)
    return model
```

```python
model_ft = train_model(model_ft, criterion, optimizer_ft, dataloaders, dataset_sizes, num_epochs=10)
torch.save(model_ft.state_dict(), 'modelo_ambiental_mobilenetv2.pth')
```

```python
%%writefile app.py
import streamlit as st
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image

# Configuración de la página
st.set_page_config(page_title="Reporte Ambiental IA", layout="centered")
st.title("🌱 Sistema Inteligente de Monitoreo Ambiental")
st.write("Sube una imagen para clasificar el estado ambiental del área y generar un plan de acción.")

@st.cache_resource
def cargar_modelo():
    model = models.mobilenet_v2(weights=None)
    num_ftrs = model.classifier[1].in_features
    model.classifier[1] = nn.Sequential(
        nn.Dropout(p=0.2, inplace=False),
        nn.Linear(num_ftrs, 3)
    )
    try:
        model.load_state_dict(torch.load('modelo_ambiental_mobilenetv2.pth', map_location=torch.device('cpu')))
        model.eval()
        return model, True
    except Exception as e:
        st.error(f"Error al cargar el modelo: {e}")
        return model, False

transformacion = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

clases = ['contaminado_agua', 'limpio', 'residuos_solidos']

soluciones = {
    'residuos_solidos': [
        "🚨 **Acción requerida:** Desplegar equipo de limpieza física al área de inmediato.",
        "📋 **Prevención:** Instalar contenedores de basura estratégicos y señalización sobre multas por tirar escombros."
    ],
    'contaminado_agua': [
        "🧪 **Acción requerida:** Enviar unidad de control ambiental para tomar muestras y evaluar toxicidad.",
        "🚧 **Prevención:** Bloquear el acceso público al cuerpo de agua temporalmente y rastrear el origen del vertido (drenajes, químicos)."
    ],
    'limpio': [
        "✅ **Acción requerida:** Ninguna inmediata. El ecosistema se encuentra estable.",
        "👁️ **Prevención:** Mantener el programa de monitoreo regular y rondas de vigilancia preventiva."
    ]
}

modelo, modelo_cargado = cargar_modelo()

if not modelo_cargado:
    st.warning("⚠️ Sube tu archivo 'modelo_ambiental_mobilenetv2.pth' a Colab para que haga predicciones reales.")

imagen_subida = st.file_uploader("Selecciona una imagen (JPG, PNG)", type=["jpg", "jpeg", "png"])

if imagen_subida is not None:
    image = Image.open(imagen_subida).convert('RGB')
    st.image(image, caption='Imagen cargada', width='stretch')

    if st.button("Analizar Imagen") and modelo_cargado:
        with st.spinner('Analizando con IA...'):
            img_tensor = transformacion(image).unsqueeze(0)
            with torch.no_grad():
                outputs = modelo(img_tensor)
                probabilidades = torch.nn.functional.softmax(outputs[0], dim=0)
                confianza, prediccion = torch.max(probabilidades, 0)

            categoria_detectada = clases[prediccion.item()]
            porcentaje = confianza.item() * 100

            st.success("Análisis completado")

            col1, col2 = st.columns(2)
            with col1:
                st.subheader(f"Detección: {categoria_detectada.upper().replace('_', ' ')}")
            with col2:
                st.metric(label="Confianza del Modelo", value=f"{porcentaje:.2f}%")

            st.progress(confianza.item())

            st.divider()

            st.subheader("🛠️ Plan de Acción Recomendado:")
            for accion in soluciones[categoria_detectada]:
                st.info(accion)
```
